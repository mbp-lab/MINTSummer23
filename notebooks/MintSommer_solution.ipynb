## 1.2) if - else Anweisung

Jetzt weißt du schon was Variablen sind und wie man ihnen Werte zuschreibt und kennst den print-Befehl. Jetzt lernen wir was **if** und **else** bedeutet.

Wenn die Bedingung hinter dem if (deutsch: wenn) erfüllt ist, also richtig ist, dann wird der Befehl hinter dem Doppelpunkt ausgeführt. Falls die Bedingung nicht erfüllt ist wird der Befehl hinter dem else (deutsch: dann) ausgeführt.

In dem Beispiel unten wird verglichen ob das Alter kleiner ist als 18. Falls diese Bedingung erfüllt ist dann wird "Du bist noch nicht volljährig" ausgegeben. Falls die Bedingung nicht erfüllt ist wird der else-Fall ausgeführt und "Du bist volljährig" ausgegeben.

```python
# Trage hier dein Alter ein.
alter =  23

if alter < 18:
  print("Du bist noch nicht volljährig!")
else:
  print("Du bist volljährig")
```

## 1.3) for Schleife

Eine sogennante **for-Schleife** (engl: for-loop) wird mehrfach ausgeführt. Hier wird die Schleife für jeden Eintrag in der Liste *hobbies* ausgeführt. Das passiert durch das Wort **in**. Somit wird durch die *hobbies* Liste iterriert (fancy Informatikwort für wiederholter Durchlauf) und das jeweilige Wort in der Liste wird in der Variable *hobby* gespeichert. Dann wird der print-Befehl ausgeführt und dann wird die Variable *hobby* wieder überschrieben mit dem nächsten Wort der Liste. Die Schleife terminiert (endet) wenn alle Werte aus der Liste einmal dran waren.

```python
# Trage hier deine Hobbies ein (zB  ["singen", "kochen"])
hobbies = ["Ski fahren", "schwimmen", "backen"]

for hobby in hobbies:
  print("Ich mag gerne:", hobby)
```

Das Facial Action Coding System (FACS) ist ein System zur Taxonomisierung menschlicher Gesichtsbewegungen. Die Bewegungen der einzelnen Gesichtsmuskeln werden von FACS anhand von geringfügig unterschiedlichen momentanen Veränderungen des Gesichtsausdrucks kodiert. Mit Hilfe von FACS ist es möglich, nahezu jeden anatomisch möglichen Gesichtsausdruck zu kodieren und ihn in die spezifischen Aktionseinheiten (AU) zu zerlegen, die den Ausdruck erzeugt haben. Es ist ein gängiger Standard zur objektiven Beschreibung von Gesichtsausdrücken [[1] ](https://github.com/TadasBaltrusaitis/OpenFace/wiki/Action-Units).



## 2.2) Wie sehen FAC aus?

Wir wollen uns nun die Facial Action Units für das Video oben angucken. Die Facial Action Units sind bereits extrahiert und in einer csv-Datei gespeichert. Eine csv-Datei ist eine Tabelle, wie zum Beispiel in Excel. Unten könnt ihr einen Ausschnitt dieser Tabelle sehen.



```python
# Lade die csv-Datei (Tabelle) mit den Facial Action Unit Ergebnissen für das Video oben
df = pd.read_csv("MINTSummer23/HermineRon.csv")

#code adapted from https://towardsdatascience.com/how-to-extract-facial-expressions-head-pose-and-gaze-from-any-youtube-video-2aa6590c2bb6
# Threshold data by 80%
df_clean = df[df.confidence>=.80]

# Define a regex pattern for all Actio Units
au_regex_pat = re.compile(r'^AU[0-9]+_[rc]$')

# Select the desired columns based on the combined regex pattern
selected_columns = ['frame', 'timestamp', 'confidence', 'success'] + \
                   df_clean.columns[df_clean.columns.str.contains(au_regex_pat)].tolist()

# Show the head of the dataframe with the selected columns
df_clean[selected_columns].head()
```

 In der Tabelle seht ihr neben dem Frame und dem Zeitstempel die Werte für alle Facial Action Units. Ihr seht das diese entweder _r oder _c am Ende des Namens haben. Alle Werte die _r am Ende haben beschreiben die Intensität der FAU in einem Interval von 0-1, während _c die Präsenz der FAU beschreibt. Das bedeutet wenn die FAU vorhanden ist, ist der Wert 1.0 sonst 0.0. Wir wollen uns nun die Intensität etwas genauer ansehen.
 
 Unten seht ihr Graphen für die Intensität der verschiedenen Action Units. Auf der y-Achse ist die Intensität der jeweiligen FAC abgezeichnet, während die x-Achse die frames anzeigt. Frames sind die einzelnen Bilder in einem Video, also man könnte auch sagen das auf der x-Achse die Zeit aufgetragen ist.

```python
# Plot all Action Unit time series. 
au_regex_pat = re.compile(r'^AU[0-9]+_r$')
au_columns_r = df_clean.columns[df_clean.columns.str.contains(au_regex_pat)]
print("List of AU columns:", au_columns_r)
f,axes = plt.subplots(6, 3, figsize=(10,12), sharex=True, sharey=True)
axes = axes.flatten()
for au_ix, au_col in enumerate(au_columns_r):
    plt.sca(axes[au_ix])
    plt.plot(df_clean['frame'].values, df_clean[au_col].values, label=' ')
    plt.xlabel('Frame')
    plt.ylabel('Intensity')
    plt.title(au_col)
    plt.legend()
    
plt.suptitle("AU intensity predictions", y=1.02)
plt.tight_layout()
plt.show()
```

**c)** Merkel blinzelt im Video einige Male. Facial Action Unit 45 zeigt den Lidschlag an. Zähle anhand der csv-Datei die Anzahl der Vorkomnisse und kontrolliere dein Ergebnis am Video. 

Hinweis: Ein Blinzeln fängt mit Präsenz 0.0 an (Augen offen), danach ist Präsenz 1.0 (Augen zu) und dann wieder offen. Somit ist die Anzahl der Frames, bei welchen die Augen zu sind nicht die Anzahl an Blinzeln.



```python
# initialisiere eine Zählvariable mit 0
count = 0

# gehe über alle Zeilen bzw Frames des Datframes
for i in range(1, len(df)+1):
  # value ist der Wert (1.0 oder 0.0) im aktuellen frame i
  value = df.loc[df['frame']==i, 'AU45_c'].values[0]
  # für den ersten Frame gibt es keinen vorherigen Frame, also müssen wir das vorher prüfen
  if i != 1:
    # prev_value ist der Wert im vorherigen Frame
    prev_value = df.loc[df['frame']==i-1, 'AU45_c'].values[0]
    # falls die Augen jetzt geschlossen (1.0) sind und davor offen (0.0)
    if value == 1.0 and prev_value == 0.0:
      # wird die Zählvariable um eins hochgezählt
      count += 1
  # falls die Augen schon im ersten Frame geschlossen sind, zählen wir das als Blinzeln
  else:
    if value == 1.0:
      count += 1
      
print("Merkel blinzelt im Video", count, "mal")
```

```python
# Eine andere Lösung für die Aufgabe

# verwende shift um den Wert in der vorherigen Zeile zu erhalten
prev_row = df['AU45_c'].shift(1)

# zähle wie of, AU45_c gleich 1,0 und die vorherige Zeile gleich 0,0 ist
count = len(df[(df['AU45_c'] == 1.0) & (prev_row == 0.0)])

print("Merkel blinzelt im Video", count, "mal")
```
